```python
# Cell 1 – Imports
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML, display
import math
import os
```

```python
# Cell 2 – Parameters
target_x = 80
target_y = 50
target_radius = 5
danger_zone_radius = 15
frames = 200
interval = 100
```

```python
# Cell 3 – Helper Functions
def drone_position(frame):
    # Parametric spiral equation for drone trajectory
    t = frame / 20.0
    x = 20 + 50 * math.exp(-t / 5.0) * math.cos(t)
    y = 50 + 30 * math.exp(-t / 5.0) * math.sin(t)
    return x, y

def euclidean_distance(x1, y1, x2, y2):
    # Calculates the physical distance between two points
    return math.sqrt((x1 - x2)**2 + (y1 - y2)**2)
```

```python
# Cell 4 – Visualization Setup
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)
ax.set_aspect('equal')
ax.grid(True)
ax.set_xlabel('X Coordinate')
ax.set_ylabel('Y Coordinate')
ax.set_title('Drone Approaching - Monitoring')

# Visualize protected target
target_circle = plt.Circle((target_x, target_y), target_radius, color='green', label='Target')
ax.add_patch(target_circle)

# Visualize danger/detection zone
danger_zone_circle = plt.Circle((target_x, target_y), danger_zone_radius, color='red', alpha=0.3, label='Danger / Detection Zone')
ax.add_patch(danger_zone_circle)

# Initialize drone representation at starting point
drone_marker, = ax.plot([], [], marker='o', color='blue', linestyle='None', label='Drone')

ax.legend(loc='upper left')

# Global state to prevent processing after interception
detected = False
```

```python
# Cell 5 – Animation Update Logic
print("\U0001f681 Drone simulation started...")

def update(frame):
    global detected
    
    # Process movement only if drone hasn't been intercepted
    if not detected:
        x, y = drone_position(frame)
        drone_marker.set_data([x], [y])
        
        distance = euclidean_distance(x, y, target_x, target_y)
        
        # Detection occurs upon entering the danger zone
        if distance < danger_zone_radius:
            detected = True
            danger_zone_circle.set_alpha(1.0)
            ax.set_title('🚨 DRONE ATTACK DETECTED! (RED ALERT)')
            drone_marker.set_color('black')  # Indicates interception
            
            print(f"\u2705 Drone detected at frame {frame} (distance = {distance:.3f})")
            print("\ud83d� Drone intercepted!")
            
            # Save visual proof
            fig.savefig('detected_moment.png', dpi=150, bbox_inches='tight')
            
    # Actions for the final execution frame
    if frame == frames - 1:
        fig.savefig('simulation_final.png', dpi=150, bbox_inches='tight')
        print("\u2705 Simulation finished.")
        print(f"File saved: {os.path.abspath('detected_moment.png')}")
        print(f"File saved: {os.path.abspath('simulation_final.png')}")
        
    return drone_marker, danger_zone_circle
```

```python
# Cell 6 – Run and Display Animation
ani = animation.FuncAnimation(fig, update, frames=frames, interval=interval, blit=False)
display(HTML(ani.to_jshtml()))
plt.close()
```
